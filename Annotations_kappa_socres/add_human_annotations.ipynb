```python
import pandas as pd

human_annotations = pd.read_csv('../Annotated_datasets/human_annotation_occupation.csv')

add_label_male = [
    "constructor",
    "mechanician",
    "carpenter",
    "driver",
    "sheriff",
    "developer",
    "guard",
    "mover",
    "chief",
    "farmer",
    "CEO",
    "janitor",
    "analyst",
    "manager",
    "lawyer",
    "cook",
    "supervisor",
    "physician",
    "salesperson",
    "laborer",
    "worker"
]

occupations_to_add = [
    "accountant",
    "analyst",
    "assistant",
    "attendant",
    "auditor",
    "baker",
    "carpenter",
    "cashier",
    "CEO",
    "chief",
    "cleaner",
    "clerk",
    "constructor",
    "cook",
    "counselor",
    "designer",
    "developer",
    "driver",
    "editor",
    "farmer",
    "guard",
    "hairdresser",
    "housekeeper",
    "janitor",
    "laborer",
    "lawyer",
    "librarian",
    "manager",
    "mechanic",
    "mover",
    "nurse",
    "physician",
    "receptionist",
    "salesperson",
    "secretary",
    "tailor",
    "sheriff",
    "supervisor",
    "teacher",
    "writer",
    "worker"
]

add_label_female = set(occupations_to_add) - set(add_label_male)
```

```python
dataset = pd.read_csv('../Annotated_datasets/Winobias_annotated.csv')

def get_gender_label_annotation_1(occupation):
    human_annotations = pd.read_csv('../Annotated_datasets/human_annotation_occupation.csv')

    human_annotations1 = human_annotations["Annotator 1"].tolist()
    human_annotations2 = human_annotations["Annotator 2"].tolist()
    corresponding_occupations = human_annotations["occupation"].tolist()

    annotations_1 = {}
    annotations_2 = {}

    for i in range(len(human_annotations1)):
        annotations_1[corresponding_occupations[i]] = human_annotations1[i]
        annotations_2[corresponding_occupations[i]] = human_annotations2[i]
    if occupation in annotations_1 and occupation in annotations_2:
        if annotations_1[occupation] == "f":
            return 0
        else:
            return 1
    else:
        return None

def get_gender_label_annotation_2(occupation):
    human_annotations = pd.read_csv('../Annotated_datasets/human_annotation_occupation.csv')

    human_annotations1 = human_annotations["Annotator 1"].tolist()
    human_annotations2 = human_annotations["Annotator 2"].tolist()
    corresponding_occupations = human_annotations["occupation"].tolist()

    annotations_1 = {}
    annotations_2 = {}

    for i in range(len(human_annotations1)):
        annotations_1[corresponding_occupations[i]] = human_annotations1[i]
        annotations_2[corresponding_occupations[i]] = human_annotations2[i]
    if occupation in annotations_1 and occupation in annotations_2:
        if annotations_2[occupation] == "f":
            return 0
        else:
            return 1
    else:
        return None# Unknown occupation

# Add the new column
dataset['annotator_1'] = dataset['context'].apply(get_gender_label_annotation_1)
dataset['annotator_2'] = dataset['context'].apply(get_gender_label_annotation_2)

dataset['annotator_1'] = dataset['annotator_1'].astype('Int64')
dataset['annotator_2'] = dataset['annotator_2'].astype('Int64')
# Save the updated dataset
dataset.to_csv('winobias_occ', index=False)

print("Added gender_US_statistics column to the dataset")
print(f"Total rows: {len(dataset)}")
print(f"Male-labeled (1): {(dataset['annotator_1'] == 1).sum()}")
print(f"Female-labeled (0): {(dataset['annotator_1'] == 0).sum()}")
print(f"Unknown: {dataset['annotator_1'].isna().sum()}")
print("\nSample of the data:")
print(dataset[['context', 'annotator_1']].head(20))


print("--------------------------------------------------------------")

print("Added gender_US_statistics column to the dataset")
print(f"Total rows: {len(dataset)}")
print(f"Male-labeled (1): {(dataset['annotator_2'] == 1).sum()}")
print(f"Female-labeled (0): {(dataset['annotator_2'] == 0).sum()}")
print(f"Unknown: {dataset['annotator_2'].isna().sum()}")
print("\nSample of the data:")
print(dataset[['context', 'annotator_2']].head(20))

```

```python
import pandas as pd

dataset = pd.read_csv('./GenderLex_occ_annotated - GenderLex_occ.csv')

def change_gender_label(label):
    if label == "M":
        return 1
    elif label == "W":
        return 0
    else:
        return None
dataset['annotator 2'] = dataset['annotator 2'].apply(change_gender_label)

dataset.to_csv('GenderLex_annotated', index=False)

print("Added gender_US_statistics column to the dataset")
print(f"Total rows: {len(dataset)}")
print(f"Male-labeled (1): {(dataset['annotator 2'] == 1).sum()}")
print(f"Female-labeled (0): {(dataset['annotator 2'] == 0).sum()}")
print(f"Unknown: {dataset['annotator 2'].isna().sum()}")
print("\nSample of the data:")
print(dataset[['context', 'annotator 2']].head(20))
```

```python
dataset = pd.read_csv('../Annotated_datasets/Winobias_annotated.csv')



def get_gender_label(occupation):
    if occupation in add_label_male:
        return 1  # Male
    elif occupation in add_label_female:
        return 0  # Female
    else:
        return None  # Unknown occupation

# Add the new column
dataset['gender_US_statistics'] = dataset['context'].apply(get_gender_label)

dataset['gender_US_statistics'] = dataset['gender_US_statistics'].astype('Int64')

dataset.to_csv('winobias_occ', index=False)

print("Added gender_US_statistics column to the dataset")
print(f"Total rows: {len(dataset)}")
print(f"Male-labeled (1): {(dataset['gender_US_statistics'] == 1).sum()}")
print(f"Female-labeled (0): {(dataset['gender_US_statistics'] == 0).sum()}")
print(f"Unknown: {dataset['gender_US_statistics'].isna().sum()}")
print("\nSample of the data:")
print(dataset[['context', 'gender_US_statistics']].head(20))
```
